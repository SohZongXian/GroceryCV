## Imports

```python
import os
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D
from tensorflow.keras.losses import sparse_categorical_crossentropy, binary_crossentropy
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
```

## Datagen object creation

```python
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
##image aug datagen
datagen_aug = tf.keras.preprocessing.image.ImageDataGenerator(
    rescale = 1./255., # Rescaling
    rotation_range = 40, # for augmentation
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)

##normal datagen
datagen = ImageDataGenerator(rescale = 1./255.) # Rescaling
```

## Loading image with augmentation

```python
image_size = (224, 224)
```

```python
##augmented datagen
training_set = datagen_aug.flow_from_directory('C:/Users/user/Documents/GitHub/GroceryCV/GroceryStoreDataset-master/dataset/train',
                                          batch_size=32,
                                           target_size = image_size,
                                          class_mode='categorical')
```

## Loading image without augmentation 

```python
##non augmented datagen
training_set = datagen.flow_from_directory('C:/Users/user/Documents/GitHub/GroceryCV/GroceryStoreDataset-master/dataset/train',
                                          batch_size=32,
                                           target_size = image_size,
                                          class_mode='categorical')
```

```python
##plot images
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
    plt.tight_layout()
    plt.show()
    
    
augmented_images = [training_set[0][0][0] for i in range(10)]
plotImages(augmented_images)
```

## Load test data

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('C:/Users/user/Documents/GitHub/GroceryCV/GroceryStoreDataset-master/dataset/test',
                                          batch_size=32,
                                        target_size = image_size,
                                          class_mode='categorical')
```

```python
augmented_images = [test_set[0][0][0] for i in range(10)]
plotImages(augmented_images)
```

```python
labels = (training_set.class_indices)
labels
```

## Build Model

```python
# Initialising the CNN
cnn = Sequential()
cnn.add(Conv2D(16, kernel_size=(5, 5), activation='relu', input_shape=(224, 224, 3)))
cnn.add(Conv2D(32, kernel_size=(5, 5), activation='relu'))
cnn.add(Conv2D(64, kernel_size=(5, 5), activation='relu'))
# cnn.add(Conv2D(64, kernel_size=(5, 5), activation='relu'))
cnn.add(Flatten())
cnn.add(Dense(16, activation='relu'))
cnn.add(Dense(len(labels), activation='softmax'))

# Display a model summary
cnn.summary()

cnn.compile(loss='categorical_crossentropy',
              optimizer=Adam(),
              metrics=['accuracy']
             )
```

```python
history = cnn.fit(x = training_set, validation_data = test_set, epochs = 15)
```
